# Class "TWOSTACK"

```python
class TWOSTACK:
    # Constructor
    def __init__(self, size1, size2):
        self.size1 = size1
        self.size2 = size2
        self.arr = [None for i in range(size1+size2)]
        self.top1 = -1
        self.top2 = -1
    
    # Check if Stack1 is full
    def isFull1(self):
        return self.top1 == self.size1 - 1
    
    # Check if Stack2 is full
    def isFull2(self):
        return self.top2 == self.size2 - 1
    
    # Check if Stack1 is empty
    def isEmpty1(self):
        return self.top1 == -1
    
    # Check if Stack2 is empty
    def isEmpty2(self):
        return self.top2 == -1
    
    # Push an element x to Stack1
    def push1(self, x):
        if self.top1 < self.size1 - 1:
            self.top1 = self.top1 + 1
            self.arr[self.top1] = x
            return True
        else:
            print("Stack1 Overflow")
            return False
    
    # Push an element x to Stack2
    def push2(self, x):
        if self.top2 < self.size2 - 1:
            self.top2 = self.top2 + 1
            self.arr[self.size1 + self.top2] = x
            return True
        else:
            print("Stack2 Overflow")
            return False
    
    # Pop the last element from Stack1
    def pop1(self):
        if self.top1 > -1:
            x = self.arr[self.top1]
            self.arr[self.top1] = None
            self.top1 = self.top1 - 1
            return x
        else:
            print("Stack1 Underflow")
            return False
    
    # Pop the last element from Stack2
    def pop2(self):
        if self.top2 > -1:
            x = self.arr[self.size1 + self.top2]
            self.arr[self.size1 + self.top2] = None
            self.top2 = self.top2 - 1
            return x
        else:
            print("Stack2 Underflow")
            return False
    
    # Peek the last element from Stack1
    def peek1(self):
        if self.top1 > -1:
            x = self.arr[self.top1]
            return x
        else:
            print("Stack1 Empty")
            return False
    
    # Peek the last element from Stack2
    def peek2(self):
        if self.top2 > -1:
            x = self.arr[self.size1 + self.top2]
            return x
        else:
            print("Stack2 Empty")
            return False
    
    def display1(self):
        tmpArr = [None for i in range(self.size1)]
        for i in range(self.size1):
            tmpArr[i] = self.arr[i]
        print(tmpArr)
```

# Class "QUEUE"

```python
class QUEUE:
    def __init__(self, maxsize):
        self.stack = TWOSTACK(maxsize,maxsize)
    
    def isFull(self):
        return self.stack.isFull1()
    
    def isEmpty(self):
        return self.stack.isEmpty1()
    
    def enqueue(self, item):
        if self.stack.isFull1():
            print("Queue is full!")
            return False
        else:
            self.stack.push1(item)
            return True
    
    def dequeue(self):
        if self.stack.isEmpty1():
            print("Queue is empty!")
            return False
        else:
            while not(self.stack.isEmpty1()):
                self.stack.push2(self.stack.pop1())
            x = self.stack.pop2()
            while not(self.stack.isEmpty2()):
                self.stack.push1(self.stack.pop2())
            return x
    
    def display(self):
        self.stack.display1()
    
    def size(self):
        return (self.stack.top1 + 1)
```
